```python
from langchain_community.document_loaders import OnlinePDFLoader

# URL of the PDF you want to load
pdf_url = "https://info.publicintelligence.net/MCIA-EthiopiaCultureGuide.pdf"

# --- Step 1: Load the PDF directly from the URL ---
# The OnlinePDFLoader handles the downloading and loading in one step.
print(f"Loading PDF from URL: {pdf_url}")
loader = OnlinePDFLoader(pdf_url)
```

```python
# The .load() method fetches and parses the PDF.
documents = loader.load()
```

```python
# Note: OnlinePDFLoader will load the entire PDF as a single document
print(f"Loaded {len(documents)} document(s) from the URL.")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

print("Splitting document into smaller chunks...")
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,  # Max size of each chunk
    chunk_overlap=150   # Overlap between chunks
)
chunks = text_splitter.split_documents(documents)
print(f"Split the document into {len(chunks)} chunks.")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = FAISS.from_documents(chunks, embeddings)
print("FAISS vector store created successfully.")
```

```python
query = "What is the role of family in Ethiopian society?"
print(f"\nSearching for: '{query}'")
results = db.similarity_search(query, k=3)
```

```python
if results:
    print("\n--- Top 3 Search Results ---")
    for i, doc in enumerate(results):
        # The metadata might not have a page number if the whole PDF was one document,
        # but the source URL will be present.
        print(f"Result {i+1} (Source: {doc.metadata.get('source', 'N/A')}):")
        print(f"  Content: '{doc.page_content}'\n")
else:
    print("No relevant documents found.")
```
